```python
##Creating a dictionary of items
item = {"onion":{"fresh":1, "frozen":0, "spicy":1, "sweet":0, "green":1, "red":0, "local":1, "small":0, "expensive":0, "healthy":1},
          "cucumber":{"fresh":1, "frozen":0, "spicy":0, "sweet":1, "green":0, "red":1, "local":1, "small":0, "expensive":0, "healthy":1},
          "cabbage":{"fresh":1, "frozen":0, "spicy":0, "sweet":1, "green":1, "red":0, "local":0, "small":1, "expensive":1, "healthy":0},
          "carrot":{"fresh":0, "frozen":1, "spicy":0, "sweet":1, "green":0, "red":0, "local":1, "small":0, "expensive":0, "healthy":1},
          "jalapeno":{"fresh":1, "frozen":0, "spicy":1, "sweet":0, "green":1, "red":0, "local":0, "small":1, "expensive":1, "healthy":0}}
## defining cases of customers
case_a = {"fresh": 0.5, "spicy": 0.7, "red": 0.1}
case_b = {"green": 0.4, "frozen": 0.3, "sweet": 0.8, "expensive": 0.2}
case_c = {"fresh": 0.5, "green": 0.01, "red": 0.15, "sweet": 0.4}
case_d = {"fresh":0.8, "frozen": 0.01,}
case_e = {"fresh": 0.5, "green": 0.01,  "sweet": 0.4}
## grading and sorting
def soft(item, case):
    maks = 0
    suma = 0
    maks_name = []
    for name in item:
        for cecha in case:
            suma += case[cecha] * item[name][cecha]
        if suma > maks:
            maks = suma
            maks_name = [name]
        elif suma == maks:
            maks_name.append(name)
        suma = 0
    return maks_name

print("best choices for you are: ",soft(item, case_a))
print("best choices for you are: ",soft(item, case_b))
print("best choices for you are: ",soft(item, case_c))
print("best choices for you are: ",soft(item, case_d))
print("best choices for you are: ",soft(item, case_e))
```
